```r
data <- read.csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/labs/module_2/data/Spacex.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDS0321ENSkillsNetwork26802033-2022-01-01", header=TRUE)
```

```r
library(dplyr)
library(psych)
library(data.table)
library(magrittr)
```

```r
colnames(data)
```

```r
data %<>% 
mutate(Date= as.Date(Date, format= "%d-%m-%Y")) %<>% 
mutate(Mission_Outcome=trimws(Mission_Outcome)) %>%
mutate(Landing._Outcome=trimws(Landing._Outcome))

data %>% 
  mutate(across(where(is.character), trimws))

data
```

```r
describe(data)
```

```r
data %>% 
filter(Payload %like% "CRS") %>%
summarise(TOTAL_PAYLOAD=sum(PAYLOAD_MASS__KG_))
```

```r
data %>% 
filter(Booster_Version == "F9 v1.1") %>%
summarise(AVG_PAYLOAD=mean(PAYLOAD_MASS__KG_))
```

```r
data %>% 
filter(Landing._Outcome == "Success (ground pad)") %>%
summarise(FIRST_SUCCESS_GP=min(Date))
```

```r
data %>% 
filter(Landing._Outcome == "Success (drone ship)" & (PAYLOAD_MASS__KG_ >= 4000) & (PAYLOAD_MASS__KG_ <=7000)) %>%
summarise(dist_boot=unique(Booster_Version))
```

```r
data %>% 
group_by(Mission_Outcome) %>%
summarise(QTY=n()) %>%
arrange(desc(QTY))
```

```r
data %>% 
filter(PAYLOAD_MASS__KG_ == max(data$PAYLOAD_MASS__KG_)) %>%
summarise(Booster_Version = unique(Booster_Version)) %>%
arrange(Booster_Version)
```

```r
data %>%
filter(year(Date)==2015 & Landing._Outcome == 'Failure (drone ship)') %>%
select(Booster_Version, Launch_Site)
```

```r
data %>%
filter(between(Date, as.Date("2010-06-04", format= "%Y-%m-%d"), as.Date("2017-03-20", format= "%Y-%m-%d"))) %>%
group_by(Landing._Outcome) %>%
summarise(qty = n()) %>%
arrange(desc(qty))
```
